```python
#task3
import pygame
import sys

size = width, heigth = 1000, 600
screen = pygame.display.set_mode(size)
pygame.display.set_caption("Geometric Drawing")
clock = pygame.time.Clock()

width, height = screen.get_size()

player_pos = pygame.Vector2(screen.get_width() / 2, screen.get_height() / 2)

white = 255, 255, 255
red = 255, 0, 0

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    screen.fill(white)
    
    pygame.draw.circle(screen, red,player_pos, 25)
    
    keys = pygame.key.get_pressed()
    if keys[pygame.K_UP] and player_pos.y - 25 > 0:
        player_pos.y -= 20 
    if keys[pygame.K_DOWN] and player_pos.y + 25 < height:
        player_pos.y += 20
    if keys[pygame.K_LEFT] and player_pos.x - 25 > 0:
        player_pos.x -= 20 
    if keys[pygame.K_RIGHT] and player_pos.x + 25 < width:
        player_pos.x += 20 
    
    
        
    pygame.display.flip()
    
    dt = clock.tick(60) / 1000
    
pygame.quit()
sys.exit()
```

```python
#task2
import pygame
import os
import re
import random

pygame.init()
pygame.mixer.init()

screen = pygame.display.set_mode((600, 400))
pygame.display.set_caption("Pygame MP3 Player")

WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
YELLO = (255, 255, 0)

font = pygame.font.Font(None, 36)
small_font = pygame.font.Font(None, 28)

music_folder = r"C:\Users\Minsiz\Desktop\pygame"

songs = []
current_index = 0 
is_playing = False 

def playlist_init():
    for file in os.listdir(r"C:\Users\Minsiz\Desktop\pygame"):
        file_name = os.fsdecode(file)
        ok = re.findall("mp3$", file_name)
        if ok:
            songs.append(os.path.join(music_folder, file_name))
    if not songs:
        print("could not find song")
        exit(1)

def play_song():
    global is_playing
    pygame.mixer.music.load(songs[current_index])
    pygame.mixer.music.play()
    is_playing = True

def next_song():
    global current_index
    current_index = (current_index + 1) % len(songs)  
    play_song()

def previous_song():
    global current_index
    current_index = (current_index - 1) % len(songs)  
    play_song()

def toggle_pause():
    global is_playing
    if is_playing:
        pygame.mixer.music.pause()
    else:
        pygame.mixer.music.unpause()
    is_playing = not is_playing

def stop_music():
    pygame.mixer.music.stop()

def draw_screen():
    screen.fill(WHITE)

    if songs:
        song_text = f"now playing: {os.path.basename(songs[current_index])}"
    else:
        song_text = "no MP3 files found"

    text_surface = font.render(song_text, True, BLACK)
    screen.blit(text_surface, (20, 50))

    instructions = [
        "controls:",
        "space - play / pause",
        "right - next song",
        "left - previous song",
        "s - stop",
        "esc - exit"
    ]

    y_offset = 150
    for line in instructions:
        text_surface = small_font.render(line, True, RED)
        screen.blit(text_surface, (20, y_offset))
        y_offset += 40

    pygame.display.flip()

playlist_init()
play_song()

running = True
while running:
    draw_screen() 

    for event in pygame.event.get():
        if event.type == pygame.QUIT or (event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE):
            running = False  

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE: 
                toggle_pause()
            elif event.key == pygame.K_RIGHT: 
                next_song()
            elif event.key == pygame.K_LEFT:  
                previous_song()
            elif event.key == pygame.K_s:  
                stop_music()

pygame.quit()
```
